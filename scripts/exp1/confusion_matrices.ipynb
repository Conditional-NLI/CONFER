```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data_deberta1 = [
    [0.26, 0.47, 0.87],  # Entailment: IMPPRES, NOPE, OUR DATA
    [0.46, 0.32, 0.85],  # Neutral: IMPPRES, NOPE, OUR DATA
    [0.41, 0, 0.92],  # Contradiction: IMPPRES, NOPE, OUR DATA
]

data_roberta1= [
    [0.18, 0.47, 0.93],
    [0.50, 0.02, 0.92],
    [0.54, 0, 0.97],
]

data_deberta2= [
    [0, 0, 0.98],
    [0.56, 0.50, 0.96],
    [0, 0.45, 0.96],
]

data_roberta2= [
    [0.1, 0, 0.97],
    [0.42, 0.22, 0.95],
    [0, 0.34, 0.96],
]


datasets = ["IMPPRES", "NOPE", "CONFER"]
labels = ["Entailment", "Neutral", "Contradiction"]


df_deberta1 = pd.DataFrame(data_deberta1, index=labels, columns=labels)
df_roberta1= pd.DataFrame(data_roberta1, index=labels, columns=labels)
df_deberta2= pd.DataFrame(data_deberta2, index=labels, columns=labels)
df_roberta2= pd.DataFrame(data_roberta2, index=labels, columns=labels)


fig, axes = plt.subplots(1, 4, figsize=(28, 7), sharey=True)

# GloVe
sns.heatmap(df_deberta1, annot=True, cmap="Blues", cbar=False, ax=axes[0], linewidths=0.5)
axes[0].set_title("GloVe", fontsize=16)
#axes[0].set_xlabel("Datasets")
#axes[0].set_ylabel("Labels")

# InferSent
sns.heatmap(df_roberta1, annot=True, cmap="Blues", cbar=False, ax=axes[1], linewidths=0.5)
axes[1].set_title("InferSent", fontsize=16)
#axes[1].set_xlabel("Datasets")

# DeBERTa
sns.heatmap(df_deberta2, annot=True, cmap="Blues", cbar=False, ax=axes[2], linewidths=0.5)
axes[2].set_title("DeBERTa", fontsize=16)
#axes[2].set_xlabel("Datasets")

# RoBERTa
sns.heatmap(df_roberta2, annot=True, cmap="Blues", cbar=True, ax=axes[3], linewidths=0.5)
axes[3].set_title("RoBERTa", fontsize=16)
#axes[3].set_xlabel("Datasets")


plt.tight_layout()
#plt.savefig("models_heatmaps.pdf", format="pdf", dpi=300, bbox_inches="tight")
plt.show()
```
